```python
import numpy as np
import pandas as pd
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from algos.classical import *
from algos.advanced import *
import matplotlib.pyplot as plt
import scipy.io as sio    
import glob
from statistics import mean, stdev, variance
import time
import scipy
from sklearn.preprocessing import StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif, mutual_info_classif
import seaborn as sns
```

```python
Curve_Table = pd.read_csv('./Convergence.csv', index_col=0)
```

```python
def format_string(s):
    s = s[1:-2]
    s = s.replace('\n','')
    s = s.replace('  ',' ')
    x = s.split(' ')
    newx = [float(p) for p in x]
    return newx
```

```python
for ind in Curve_Table.index:
    
    fig, ax = plt.subplots()
    
    #BPSO
    Convergence_curve = format_string(Curve_Table['PSO'][ind])
    ax.plot(x, Convergence_curve, color='tab:blue', label = 'BPSO')
    
    
    #BBA
    Convergence_curve = format_string(Curve_Table['BA'][ind])
    ax.plot(x, Convergence_curve, color='tab:blue', label = 'BBA')
    
    #GA
    Convergence_curve = format_string(Curve_Table['GA'][ind])
    ax.plot(x, Convergence_curve, color='tab:blue', label = 'GA')
    
    #BSSA
    Convergence_curve = format_string(Curve_Table['BSSA'][ind])
    ax.plot(x, Convergence_curve, color='tab:blue', label = 'BSSA')
    
    
    #EO
    Convergence_curve = format_string(Curve_Table['EO'][ind])
    ax.plot(x, Convergence_curve, color='tab:blue', label = 'EO')
    
    
    #WOA_CM
    Convergence_curve = format_string(Curve_Table['WOA_FS'][ind])
    ax.plot(x, Convergence_curve, color='tab:blue', label = 'WOA-CM')
    
    #TMGWO
    Convergence_curve = format_string(Curve_Table['TMGWO'][ind])
    ax.plot(x, Convergence_curve, color='tab:blue', label = 'TMGWO')
    
    #TVBSSA
    Convergence_curve = format_string(Curve_Table['TVBSSA'][ind])
    ax.plot(x, Convergence_curve, color='tab:blue', label = 'TVBSSA')
    
    #Proposed
    Convergence_curve = format_string(Curve_Table['Proposed'][ind])
    ax.plot(x, Convergence_curve, color='tab:orange', label = 'IEO')
        
    ax.set_xlabel('Number of Iterations')
    ax.set_ylabel('Fitness')
    dataset = Curve_Table['Dataset'][ind]
    ax.set_title(dataset)
    ax.legend()
    ax.grid()
    plt.show()
```
